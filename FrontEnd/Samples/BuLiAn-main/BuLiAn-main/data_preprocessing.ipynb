Now we should have a look at data types. In the following we do three things:
- Remove space after team names
- Change distance from String to a float
- Change possesion from String to a float
- Change pass ratio from String to a float

```python
def remove_last_char(string):
    new_string = string[:-1]
    return new_string

def remove_km(string):
    new_string = string.replace(" km","")
    new_string_dot = new_string.replace(",",".")
    new_float = float(new_string_dot)
    return new_float

def remove_percentage(string):
    new_string = string.replace("%","")
    new_float = float(new_string)/100
    return new_float
```

```python
for i in range(len(df_data.index)):
    team_home = (df_data.at[i,'h_team'])
    team_away = (df_data.at[i,'a_team'])
    df_data.at[i,'h_team'] = remove_last_char(team_home)
    df_data.at[i,'a_team'] = remove_last_char(team_away)
```

```python
for i in range(len(df_data.index)):
    df_data.at[i,'h_distance'] = remove_km(df_data.at[i,'h_distance'])
    df_data.at[i,'a_distance'] = remove_km(df_data.at[i,'a_distance'])
    df_data.at[i,'h_pass_ratio'] = remove_percentage(df_data.at[i,'h_pass_ratio'])
    df_data.at[i,'a_pass_ratio'] = remove_percentage(df_data.at[i,'a_pass_ratio'])
    df_data.at[i,'h_possesion'] = remove_percentage(df_data.at[i,'h_possesion'])
    df_data.at[i,'a_possesion'] = remove_percentage(df_data.at[i,'a_possesion'])
    df_data.at[i,'h_tackle_ratio'] = remove_percentage(df_data.at[i,'h_tackle_ratio'])
    df_data.at[i,'a_tackle_ratio'] = remove_percentage(df_data.at[i,'a_tackle_ratio'])
```

It seems like we have issues with three games from season 13/14, where the distance was no captured. We now have three possibilities:
- Remove the whole row
- Estimate the distance based on previous values
- Manually enter the values

As I could not find these values anywhere, I chose to just average the distance covered from that season by each team for the sake of simplicity.

```python
import statistics
def avg_season_distance(season, matchday_exceptions,team):
    df_temp = df_data.loc[(df_data['season'] == season)]
    distances = []
    for i in range(len(df_temp.index)):
        if(df_temp.at[i,'h_team'] == team and df_temp.at[i,'matchday'] not in matchday_exceptions):
            distances.append(df_temp.at[i,'h_distance'])
        if(df_temp.at[i,'a_team'] == team and df_temp.at[i,'matchday'] not in matchday_exceptions):
            distances.append(df_temp.at[i,'a_distance'])
    return statistics.mean(distances)
```

```python
avg_distance_dict = dict()
avg_distance_dict["Hamburger SV"] = avg_season_distance("13-14", [4], "Hamburger SV")
avg_distance_dict["VfB Stuttgart"] = avg_season_distance("13-14", [8], "VfB Stuttgart")
avg_distance_dict["Werder Bremen"] = avg_season_distance("13-14", [8], "Werder Bremen")
avg_distance_dict["TSG Hoffenheim"] = avg_season_distance("13-14", [34], "TSG Hoffenheim")
avg_distance_dict["Eintracht Braunschweig"] = avg_season_distance("13-14", [4,34], "Eintracht Braunschweig")
```

```python
avg_distance_dict
```

```python
for i in range(len(df_data.index)):
    if(df_data.at[i,'h_distance']==0):
        print("changed data")
        df_data.at[i,'h_distance'] = avg_distance_dict[df_data.at[i,'h_team']]
    if(df_data.at[i,'a_distance']==0):
        print("changed data")
        df_data.at[i,'a_distance'] = avg_distance_dict[df_data.at[i,'a_team']]
```
